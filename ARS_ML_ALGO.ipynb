```python
data["Name_new"] = data["Name"].apply(lambda x: x.lower())
```

```python
# Remove Punctuation from a String
import string

data["Name_new"] = data["Name_new"].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
```

```python
# Removing extra spaces
data["Name_new"] = data["Name_new"].apply(lambda x: " ".join(x.split()))
```

```python
# Replacing " " with "-"
data["Name_new"] = data["Name_new"].apply(lambda x: x.replace(" ", "-"))
```

```python
def Recommender(user_input):
    Index_of_anime = data[data["Name"] == user_input].index[0]
    Similarity_score = similarity[Index_of_anime]
    Sorted_scores = sorted(list(enumerate(Similarity_score)),reverse = True,key= lambda x: x[1]) [1:6]
    Recommended_Anime = []
    
    for i in Sorted_scores:
        Recommended_Anime.append(data.iloc[i[0]].Name)
    return Recommended_Anime
        
```
